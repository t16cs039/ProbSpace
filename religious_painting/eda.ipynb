# EDA

```python
import numpy as np
import matplotlib.pyplot as plt
```

## loading data

```python
train_data = np.load('/content/drive/MyDrive/ProbSpace/宗教画テーマの分類/dataset/christ-train-imgs.npz')['arr_0']
train_label = np.load('/content/drive/MyDrive/ProbSpace/宗教画テーマの分類/dataset/christ-train-labels.npz')['arr_0']
test_data = np.load('/content/drive/MyDrive/ProbSpace/宗教画テーマの分類/dataset/christ-test-imgs.npz')['arr_0']

pb_lb = np.array([0.078, 0.078, 0.156, 0.109, 0.078, 0.047, 0.078, 0.062, 0.031, 0.078, 0.047, 0.094, 0.062])
```

## the ratio of label in train data and public LB

```python
value_count = np.unique(train_label, return_counts=True)

plt.bar(value_count[0].astype(str), pb_lb*100, alpha=0.3, color='#DF81A2', label='public LB')
plt.bar(value_count[0].astype(str), (value_count[1] / len(train_label)) * 100, alpha=0.3, color='#81D674', align='center', label='train label')

plt.title('the ratio of train label and Public LB (%)')
plt.xlabel('label')
plt.legend()

plt.show()
```
